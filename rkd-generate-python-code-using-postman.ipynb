```python
import requests
import json

url = "http://api.rkd.refinitiv.com/api/Quotes/Quotes.svc/REST/Quotes_1/RetrieveItem_3"

payload = json.dumps({
  "RetrieveItem_Request_3": {
    "ItemRequest": [
      {
        "Fields": "",
        "RequestKey": [
          {
            "Name": "MSFT.O",
            "NameType": "RIC"
          }
        ],
        "Scope": "All"
      }
    ],
    "TrimResponse": False,
    "IncludeChildItemQoS": False
  }
})
headers = {
  'X-Trkd-Auth-Token': 'RKD-API-TOKEN',
  'X-Trkd-Auth-ApplicationID': 'RKD-API-APPLICATION-ID',
  'Content-Type': 'application/json'
}

response = requests.request("POST", url, headers=headers, data=payload)

print(response.text)
```

```python
response_json = json.loads(response.text)
print(json.dumps(response_json, indent=4, sort_keys=True))
```
